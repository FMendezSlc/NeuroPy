```python
# low pass filter
fNQ = 25000 / 2               # Determine the Nyquist frequency.
n = 10                     # Define the filter order
Wn = 350 / fNQ                  # ... and specify the cutoff frequency,
b = signal.butter(n, Wn, 'low')              # ... build lowpass filter.
```

### LFP analysis

```python
channels.keys()
```

```python
def extract_lfp(channel, file_1, file_2):
    
    files = [file_1, file_2]

    for idx, file in enumerate(files):
        segment = exp_dir +'/'+ file
        reader = intanIO(filename = segment)
        reader.parse_header()
        for site in channels.keys():
            if idx == 0:
                size = reader.get_signal_size(block_index=0, seg_index=0,channel_indexes = [channel])
                raw_sig = reader.get_analogsignal_chunk(block_index=0, seg_index=0,channel_indexes= [channel], i_start = size-250000, i_stop = size-100000) # read analog_in
                sig_scaled = reader.rescale_signal_raw_to_float(raw_sig, dtype='float64', channel_indexes= [channel])
                lfp_1 = signal.filtfilt(b[0], b[1], sig_scaled.flatten(), axis = 0) # Zero-phase filter each trial
            elif idx == 1:
                size = reader.get_signal_size(block_index=0, seg_index=0,channel_indexes = [channel])
                raw_sig = reader.get_analogsignal_chunk(block_index=0, seg_index=0,channel_indexes= [channel], i_start = 100000, i_stop = 250000) # read analog_in
                sig_scaled = reader.rescale_signal_raw_to_float(raw_sig, dtype='float64', channel_indexes= [channel])
                lfp_2 = signal.filtfilt(b[0], b[1], sig_scaled.flatten(), axis = 0) # Zero-phase filter each trial
    
    lfp_signals = (lfp_1, lfp_2)
    return lfp_signals
```

```python
# Channels: channels [dic] 

file_1 = rec_files[29]
file_2 = rec_files[30]

lfps = {}
for site in channels.keys():
    lfps[site] = extract_lfp(channels[site], file_1, file_2)
```

## Spectral Analysis

```python
# Periodograms

spectral_dict = {}
for site in lfps.keys():
    fq, pw1 =  signal.welch(lfps[site][0], fs = 25000, window = 'hann', nperseg = 50000, noverlap = 25000, detrend = 'linear')
    roi_freq = fq[(fq > 0) & (fq<= 200)]
    roi_pwb = pw1[(fq > 0) & (fq<= 200)]
    
    fq, pw2 =  signal.welch(lfps[site][1], fs = 25000, window = 'hann', nperseg = 50000, noverlap = 25000, detrend = 'linear')
    roi_pwp = pw2[(fq > 0) & (fq<= 200)]
    
    spectral_dict[site] = (roi_pwb, roi_pwp)
    
```

```python
# Spectra figure

spect_fig = plt.figure(figsize= (10, 6))
ax_1= plt.subplot(2, 3, 1)
plt.plot(roi_freq, spectral_dict['DG'][0], label = 'Basal')
plt.plot(roi_freq, spectral_dict['DG'][1], label = 'Pinch')
ax_1.set_title('DG')
plt.yscale('log')
plt.xscale('log')
plt.grid(True)
plt.ylabel(r'Power ($\mu V^2$/Hz)')
plt.legend()

ax_2= plt.subplot(2, 3, 2)
plt.plot(roi_freq, spectral_dict['CA3_p'][0], label = 'Basal')
plt.plot(roi_freq, spectral_dict['CA3_p'][1], label = 'Pinch')
ax_2.set_title('CA3p')
plt.yscale('log')
plt.xscale('log')
plt.grid(True)


ax_3= plt.subplot(2, 3, 4)
plt.plot(roi_freq, spectral_dict['DG_mol'][0], label = 'Basal')
plt.plot(roi_freq, spectral_dict['DG_mol'][1], label = 'Pinch')
ax_3.set_title('DGmol')
plt.yscale('log')
plt.xscale('log')
plt.grid(True)
plt.xlabel('Frequency (Hz)')
plt.ylabel(r'Power ($\mu V^2$/Hz)')

ax_4= plt.subplot(2, 3, 5)
plt.plot(roi_freq, spectral_dict['CA3_r'][0], label = 'Basal')
plt.plot(roi_freq, spectral_dict['CA3_r'][1], label = 'Pinch')
ax_4.set_title('CA3r')
plt.yscale('log')
plt.xscale('log')
plt.grid(True)
plt.xlabel('Frequency (Hz)')

ax_5= plt.subplot(2, 3, 3)
plt.plot(roi_freq, spectral_dict['CA1_p'][0], label = 'Basal')
plt.plot(roi_freq, spectral_dict['CA1_p'][1], label = 'Pinch')
ax_5.set_title('CA1p')
plt.yscale('log')
plt.xscale('log')
plt.grid(True)


ax_6= plt.subplot(2, 3, 6)
plt.plot(roi_freq, spectral_dict['CA1_r'][0], label = 'Basal')
plt.plot(roi_freq, spectral_dict['CA1_r'][1], label = 'Pinch')
ax_6.set_title('CA1r')
plt.yscale('log')
plt.xscale('log')
plt.grid(True)
plt.xlabel('Frequency (Hz)')

plt.tight_layout()
```

```python
# Defining bands

bands_dic = {'delta': (0.5, 2), 'theta': (3, 6), 'gamma_slow': (30, 50), 'gamma_fast': (50, 100)}

power_dic = {'Date': [], 'Genotype': [], 'Sex': [], 'Tx': [], 'Condition': [], 'Site':[], 'total_pw': [], 'delta_abs': [], 'theta_abs': [],
                'gamma_slow_abs': [], 'gamma_fast_abs': [], 'delta_rel': [], 'theta_rel': [], 'gamma_slow_rel': [], 'gamma_fast_rel': []}
```

```python
f_res = fq[1]- fq[0]

for key in spectral_dict.keys():
    power_dic['Date'].append(Date)
    power_dic['Genotype'].append(Genotype)
    power_dic['Sex'].append(Sex)
    power_dic['Tx'].append(Tx)
    power_dic['Condition'].append('Basal')

    total_power = np.round(simps(spectral_dict[key][0], dx = f_res), 2)
    power_dic['Site'].append(key)
    power_dic['total_pw'].append(total_power)

    for band_key in bands_dic.keys():
        
        limits = bands_dic[band_key]
        band = np.logical_and(roi_freq >= limits[0], roi_freq <= limits[1])
        band_abs = np.round(simps(spectral_dict[key][0][band], dx=f_res), 2)
        power_dic[band_key+'_abs'].append(band_abs)
        band_rel = np.round(band_abs / total_power, 2)
        power_dic[band_key+'_rel'].append(band_rel)
        
    power_dic['Date'].append(Date)
    power_dic['Genotype'].append(Genotype)
    power_dic['Sex'].append(Sex)
    power_dic['Tx'].append(Tx)
    power_dic['Condition'].append('Pinch')
    total_power = np.round(simps(spectral_dict[key][1], dx = f_res), 2)
    power_dic['Site'].append(key)
    power_dic['total_pw'].append(total_power)

    for band_key in bands_dic.keys():

        limits = bands_dic[band_key]
        band = np.logical_and(roi_freq >= limits[0], roi_freq <= limits[1])
        band_abs = np.round(simps(spectral_dict[key][1][band], dx=f_res), 2)
        power_dic[band_key+'_abs'].append(band_abs)
        band_rel = np.round(band_abs / total_power, 2)
        power_dic[band_key+'_rel'].append(band_rel)
```

## Coherence

```python
coh_dic = {'Date': [], 'Genotype': [], 'Sex': [], 'Tx': [], 'Condition': [], 'Site':[], 'theta_coh':[], 'phase_lag':[]}
```

```python
# Compute coherence in theta band (3-6 Hz) and phase lag
for site in lfps.keys():
    coh_dic['Date'].append(Date)
    coh_dic['Genotype'].append(Genotype)
    coh_dic['Sex'].append(Sex)
    coh_dic['Tx'].append(Tx)
    coh_dic['Condition'].append('Basal')
    coh_dic['Site'].append(site)
    
    fqs_c, coh1, lags1 = spec.welch_coherence(lfps['CA3_p'][0], lfps[site][0], len_segment = 50000, overlap = 0.9, fs = 25000)
    theta_c = coh1[(fqs_c >= 3) & (fqs_c< 6)].mean()
    theta_lag = lags1[(fqs_c >= 3) & (fqs_c< 6)].mean()
    
    coh_dic['theta_coh'].append(theta_c)
    coh_dic['phase_lag'].append(theta_lag)
    
    coh_dic['Date'].append(Date)
    coh_dic['Genotype'].append(Genotype)
    coh_dic['Sex'].append(Sex)
    coh_dic['Tx'].append(Tx)
    coh_dic['Condition'].append('Pinch')
    coh_dic['Site'].append(site)
    
    fqs_c, coh2, lags2 = spec.welch_coherence(lfps['CA3_p'][1], lfps[site][1], len_segment = 50000, overlap = 0.9, fs = 25000)
    theta_c = coh2[(fqs_c >= 3) & (fqs_c< 6)].mean()
    theta_lag = lags2[(fqs_c >= 3) & (fqs_c< 6)].mean()
    
    coh_dic['theta_coh'].append(theta_c)
    coh_dic['phase_lag'].append(theta_lag)
    
    
```

```python
coh_df = pd.DataFrame(coh_dic)
coh_df
```

```python
coh_fig = plt.figure(figsize=(12, 4))
ax = plt.subplot(121)
coh_df[coh_df['Condition']== 'Basal'].plot(kind = 'bar', y = 'theta_coh', x = 'Site', position= 1, ax = ax, label = 'Basal', color = 'C0', width = 0.2)
coh_df[coh_df['Condition']== 'Pinch'].plot(kind = 'bar', y = 'theta_coh', x = 'Site', position= 0, ax=ax, label = 'Pinch', color = 'C1', width = 0.2)
plt.xlabel('')
plt.ylabel('Theta Band Coherence')
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
start = 0.03
for site in coh_df['Site'].unique():
    axin1 = coh_fig.add_axes([start, .8, 0.1, 0.1], projection = 'polar')
    lag1 = coh_df['phase_lag'][(coh_df['Site'] == site) & (coh_df['Condition'] == 'Basal')]
    lag2 = coh_df['phase_lag'][(coh_df['Site'] == site) & (coh_df['Condition'] == 'Pinch')]
    axin1.plot([0, lag1], [0, 1], color = 'C0', linewidth = 3)
    axin1.plot([0, lag2], [0, 1], color = 'C1', linewidth = 3)
    axin1.set_yticklabels([])
    axin1.set_xticklabels([r'$0\degree$','', '', '', r'$180\degree$'], fontsize = 8)
    axin1.set_theta_zero_location('N')
    axin1.tick_params(pad=0)
    start += 0.08
plt.tight_layout(rect=[0, 0, 1, .78])
```
