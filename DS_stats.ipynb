### Dentate Spikes Statistics

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import pingouin as pg
import numpy as np
```

```python
DS_df = pd.read_csv('/Users/labc02/Documents/PDCB_data/MK-project/LFP_data/Dentate_Spikes/Data/Pool_DS.csv')
```

```python
DS_df
```

```python
DS_df['Group'] = DS_df['Genotype'] + '_' + DS_df['Tx']
```

```python
Amp_fig = plt.figure(figsize=(4, 2))
ax = sns.boxplot(data = DS_df, x= 'Group', y= 'Amp', order = ['WT_Pl', 'WT_Tx05', 'WT_Tx1', 'KO_Pl', 'KO_Tx05'], palette=['royalblue', 'steelblue', 'cornflowerblue', 'forestgreen', 'darkseagreen'], width = 0.3, showmeans = True, meanprops={"marker":"+", "markeredgecolor":"k"}, showfliers = False)
sns.despine()
ax.set_ylabel('Amp $(\mu V)$')
ax.set_xticklabels(['WT\nSaline', 'WT\n0.5mg/kg', 'WT\n1mg/kg', 'KO\nSaline', 'KO\n0.5mg/kg'])
ax.set_xlabel('')
plt.tight_layout()
```

```python
pg.normality(data = DS_df, dv = 'Amp', group = 'Group')
```

```python
pg.anova(data = DS_df, dv = 'Amp', between = 'Group')
```

```python
Width_fig = plt.figure(figsize=(4, 2))
ax = sns.boxplot(data = DS_df, x= 'Group', y= 'Width', order = ['WT_Pl', 'WT_Tx05', 'WT_Tx1', 'KO_Pl', 'KO_Tx05'], palette=['royalblue', 'steelblue', 'cornflowerblue', 'forestgreen', 'darkseagreen'], width = 0.3, showmeans = True, meanprops={"marker":"+", "markeredgecolor":"k"}, showfliers = False)
sns.despine()
ax.set_ylabel('Width (ms)')
ax.set_xticklabels(['WT\nSaline', 'WT\n0.5mg/kg', 'WT\n1mg/kg', 'KO\nSaline', 'KO\n0.5mg/kg'])
ax.set_xlabel('')
plt.tight_layout()
```

```python
pg.normality(data = DS_df, dv = 'Width', group = 'Group')
```

```python
pg.anova(data = DS_df, dv = 'Width', between = 'Group')
```

```python
Int_fig = plt.figure(figsize=(4, 2))
ax = sns.boxplot(data = DS_df, x= 'Group', y= 'Int', order = ['WT_Pl', 'WT_Tx05', 'WT_Tx1', 'KO_Pl', 'KO_Tx05'], palette=['royalblue', 'steelblue', 'cornflowerblue', 'forestgreen', 'darkseagreen'], width = 0.3, showmeans = True, meanprops={"marker":"+", "markeredgecolor":"k"}, showfliers = False)
sns.despine()
ax.set_ylabel('Interval (s)')
ax.set_xticklabels(['WT\nSaline', 'WT\n0.5mg/kg', 'WT\n1mg/kg', 'KO\nSaline', 'KO\n0.5mg/kg'])
ax.set_xlabel('')
plt.tight_layout()
```

```python
pg.normality(data = DS_df, dv = 'Int', group = 'Group')
```

```python
pg.anova(data = DS_df, dv = 'Int', between = 'Group')
```

```python
pg.ttest(DS_df['Int'][DS_df['Group'] == 'KO_Pl'], DS_df['Int'][DS_df['Group'] == 'KO_Tx05'])
```

```python
pg.anova(data = DS_df[DS_df['Genotype'] == 'WT'], dv = 'Int', between = 'Tx')
```
