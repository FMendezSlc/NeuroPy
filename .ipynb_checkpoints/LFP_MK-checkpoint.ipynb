```python
from matplotlib import pyplot as plt
from neo.rawio import IntanRawIO as intanIO
from scipy import signal
import pandas as pd
import numpy as np
from scipy.integrate import simps
import os
```

```python
exp_dir = "/Volumes/Seagate Expansion Drive/Intan_Exp/100921_WT_91_3_Male_Tx1/WT_Male_Tx1_Spont_210910_175438/"
```

```python
rec_files = [file for file in os.listdir(exp_dir) if 'rhd' in file]
```

```python
channels = [14, 13, 16]
```

```python
# low pass filter
fNQ = 25000 / 2               # Determine the Nyquist frequency.
n = 10                     # Define the filter order
Wn = 350 / fNQ                  # ... and specify the cutoff frequency,
b = signal.butter(n, Wn, 'low')              # ... build lowpass filter.
```

```python
CA1 = np.zeros((500, 450))
CA3 = np.zeros((500, 450))
DG = np.zeros((500, 450))

col = 0
for file in rec_files[0:30]:
    segment = exp_dir + file
    reader = intanIO(filename = segment)
    reader.parse_header()
    size = reader.get_signal_size(block_index=0, seg_index=0,channel_indexes = [0])
    windows = size//(25000*4)
    
    for channel in channels:
        raw_sig = reader.get_analogsignal_chunk(block_index=0, seg_index=0,channel_indexes=[channel], i_start = 0, i_stop = size) # read analog_in
        sig_scaled = reader.rescale_signal_raw_to_float(raw_sig, dtype='float64', channel_indexes= [channel])
        lfp = signal.filtfilt(b[0], b[1], sig_scaled.flatten(), axis = 0) # Zero-phase filter each trial
        
        for window in range(windows):
            win = (window *(4 *25000), (window+1)*(4*25000))
            fx, pxx =  signal.welch(lfp[win[0]:win[1]], fs = 25000, window = 'hann', nperseg = 50000)
            roi_freq = fx[(fx > 0) & (fx<= 250)]
            roi_pxx = pxx[(fx > 0) & (fx<= 250)]
            
            if channel == 14:
                DG[:, window+ col] = roi_pxx
            elif channel == 13:
                CA1[:, window+col] = roi_pxx
            elif channel == 16:
                CA3[:, window+col] = roi_pxx
    col += 15

Ca1_cor = np.corrcoef(CA1)
np.fill_diagonal(Ca1_cor, 0.0)

Ca3_cor = np.corrcoef(CA3)
np.fill_diagonal(Ca3_cor, 0.0)

Dg_cor = np.corrcoef(DG)
np.fill_diagonal(Dg_cor, 0.0)

fund_fqs = plt.figure(figsize= (14, 8))

plt.subplot(131)
plt.imshow(Ca1_cor, origin = 'lower', cmap = 'bwr', vmin = -1, vmax = 1)
ticks = np.arange(0, 500, 50)
ticks_labels = np.arange(0.0, 250, 25)
plt.xticks(ticks, ticks_labels, rotation = 45)
plt.yticks(ticks, ticks_labels)
plt.title('CA1')

plt.subplot(132)
plt.imshow(Ca3_cor, origin = 'lower', cmap = 'bwr', vmin = -1, vmax = 1)
plt.xticks(ticks, ticks_labels, rotation = 45)
plt.yticks(ticks, ticks_labels)
plt.title('CA3')

plt.subplot(133)
plt.imshow(Dg_cor, origin = 'lower', cmap = 'bwr', vmin = -1, vmax = 1)
plt.xticks(ticks, ticks_labels, rotation = 45)
plt.yticks(ticks, ticks_labels)
plt.title('DG')


all_time = plt.figure(figsize = (12, 6))

plt.plot(np.arange(0, 1800/60, 4/60), CA1[1:200, :].mean(0), label = 'CA1')
plt.plot(np.arange(0, 1800/60, 4/60), CA3[1:200, :].mean(0), label = 'CA3')
plt.plot(np.arange(0, 1800/60, 4/60), DG[1:200, :].mean(0), label = 'DG')
plt.xlabel('min')
plt.ylabel('Power V**2/Hz')
plt.legend()
```

```python
CA1_pool = np.zeros((500, 500, 4))
CA3_pool = np.zeros((500, 500, 4))
DG_pool = np.zeros((500, 500, 4))
```

```python
CA1_pool[:, :, 0] = Ca1_cor
CA3_pool[:, :, 0] = Ca3_cor
DG_pool[:, :, 0] = Dg_cor
```

```python
CA1_mean = CA1_pool.mean(-1)
CA3_mean = CA3_pool.mean(-1)
DG_mean = DG_pool.mean(-1)
```

```python
plt.figure(figsize= (14, 8))

plt.subplot(131)
plt.imshow(CA1_mean, origin = 'lower', cmap = 'bwr', vmin = -1, vmax = 1)
ticks = np.arange(0, 500, 50)
ticks_labels = np.arange(0.0, 250, 25)
plt.xticks(ticks, ticks_labels, rotation = 45)
plt.yticks(ticks, ticks_labels)
plt.title('CA1')

plt.subplot(132)
plt.imshow(CA3_mean, origin = 'lower', cmap = 'bwr', vmin = -1, vmax = 1)
plt.xticks(ticks, ticks_labels, rotation = 45)
plt.yticks(ticks, ticks_labels)
plt.title('CA3')

plt.subplot(133)
plt.imshow(DG_mean, origin = 'lower', cmap = 'bwr', vmin = -1, vmax = 1)
plt.xticks(ticks, ticks_labels, rotation = 45)
plt.yticks(ticks, ticks_labels)
plt.title('DG')
```

### LFP analysis

```python
# Channels are: DG-14; CA3-16; CA1-13

file_1 = rec_files[29]
file_2 = rec_files[30]
files = [file_1, file_2]

for idx, file in enumerate(files):
    segment = exp_dir + file
    reader = intanIO(filename = segment)
    reader.parse_header()
    if idx == 0:
        size = reader.get_signal_size(block_index=0, seg_index=0,channel_indexes = [1])
        raw_sig = reader.get_analogsignal_chunk(block_index=0, seg_index=0,channel_indexes=[1], i_start = size-250000, i_stop = size) # read analog_in
        sig_scaled = reader.rescale_signal_raw_to_float(raw_sig, dtype='float64', channel_indexes= [1])
        lfp_1 = signal.filtfilt(b[0], b[1], sig_scaled.flatten(), axis = 0) # Zero-phase filter each trial
    elif idx == 1:
        size = reader.get_signal_size(block_index=0, seg_index=0,channel_indexes = [1])
        raw_sig = reader.get_analogsignal_chunk(block_index=0, seg_index=0,channel_indexes=[1], i_start = 0, i_stop = 250000) # read analog_in
        sig_scaled = reader.rescale_signal_raw_to_float(raw_sig, dtype='float64', channel_indexes= [1])
        lfp_2 = signal.filtfilt(b[0], b[1], sig_scaled.flatten(), axis = 0) # Zero-phase filter each trial

lfp = np.hstack((lfp_1, lfp_2))
plt.figure(figsize=(12, 6))
plt.subplot(211)
plt.plot(lfp)
plt.axis('off')
plt.subplot(212)
plt.specgram(x = lfp, Fs = 25000, NFFT = 50000, noverlap = 25000, detrend = 'mean', cmap = 'jet', vmin = 0)
plt.ylim((0.0, 30))
plt.minorticks_on()
plt.ylabel('Frequency (Hz)')
plt.tick_params(which = 'both', right = True, labelright=True)
plt.xlabel('Time')
```

```python
CA1_welch = np.zeros((500, 2))
CA3_welch = np.zeros((500, 2))
DG_welch = np.zeros((500, 2))


for channel in channels:

    for idx, file in enumerate(files):
        segment = exp_dir + file
        reader = intanIO(filename = segment)
        reader.parse_header()
        if idx == 0:
            size = reader.get_signal_size(block_index=0, seg_index=0,channel_indexes = [channel])
            raw_sig = reader.get_analogsignal_chunk(block_index=0, seg_index=0,channel_indexes=[channel], i_start = size-150000, i_stop = size-25000) # read analog_in
            sig_scaled = reader.rescale_signal_raw_to_float(raw_sig, dtype='float64', channel_indexes= [channel])
            lfp_1 = signal.filtfilt(b[0], b[1], sig_scaled.flatten(), axis = 0) # Zero-phase filter each trial
        
            fx, pxx =  signal.welch(lfp_1, fs = 25000, window = 'hann', nperseg = 50000, detrend = 'linear')
            roi_freq = fx[(fx > 0) & (fx<= 250)]
            roi_pxx = pxx[(fx > 0) & (fx<= 250)]
            
            if channel == 14:
                DG_welch[:, 0] = roi_pxx
            elif channel == 13:
                CA1_welch[:, 0] = roi_pxx
            elif channel == 16:
                CA3_welch[:, 0] = roi_pxx
        elif idx == 1:
            size = reader.get_signal_size(block_index=0, seg_index=0,channel_indexes = [channel])
            raw_sig = reader.get_analogsignal_chunk(block_index=0, seg_index=0,channel_indexes=[channel], i_start =100000, i_stop = 225000) # read analog_in
            sig_scaled = reader.rescale_signal_raw_to_float(raw_sig, dtype='float64', channel_indexes= [channel])
            lfp_2 = signal.filtfilt(b[0], b[1], sig_scaled.flatten(), axis = 0) # Zero-phase filter each trial

            fx, pxx =  signal.welch(lfp_2, fs = 25000, window = 'hann', nperseg = 50000, detrend = 'linear')
            roi_freq = fx[(fx > 0) & (fx<= 250)]
            roi_pxx = pxx[(fx > 0) & (fx<= 250)]
            
            if channel == 14:
                DG_welch[:, 1] = roi_pxx
            elif channel == 13:
                CA1_welch[:, 1] = roi_pxx
            elif channel == 16:
                CA3_welch[:, 1] = roi_pxx

lfp_welch = plt.figure(figsize= (10, 8))
gs = lfp_welch.add_gridspec(5, 3)
lfp_welch.add_subplot(gs[0, :])
plt.plot(lfp_1, 'k')
plt.axis('off')

lfp_welch.add_subplot(gs[1, :])
plt.plot(lfp_2, 'k')
plt.axis('off')

lfp_welch.add_subplot(gs[2:4, 0])
plt.plot(roi_freq, CA1_welch[:, 0], label = 'Basal')
plt.plot(roi_freq, CA1_welch[:, 1], label = 'Pinch')
plt.yscale('log')
plt.xscale('log')
plt.grid(True)
plt.legend()
plt.title('CA1')

lfp_welch.add_subplot(gs[2:4, 1])
plt.plot(roi_freq, CA3_welch[:, 0])
plt.plot(roi_freq, CA3_welch[:, 1])
plt.yscale('log')
plt.xscale('log')
plt.grid(True)
plt.title('CA3')

lfp_welch.add_subplot(gs[2:4, 2])
plt.plot(roi_freq, DG_welch[:, 0])
plt.plot(roi_freq, DG_welch[:, 1])
plt.yscale('log')
plt.xscale('log')
plt.grid(True)
plt.title('DG')

plt.tight_layout()
```
