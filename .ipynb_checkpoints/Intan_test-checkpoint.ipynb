### Filtering
#### Low-pass

```python
fNQ = sampling_rate / 2               # Determine the Nyquist frequency.
n = 10                     # Define the filter order
Wn = 250 / fNQ                  # ... and specify the cutoff frequency,
b = signal.butter(n, Wn, 'low')              # ... build lowpass filter.
lfp = signal.filtfilt(b[0], b[1], float_sigs, axis = 0) # Zero-phase filter each trial
```

#### High-pass

```python
fNQ = sampling_rate / 2               # Determine the Nyquist frequency.
n = 10                     # Define the filter order
Wn = 350 / fNQ                  # ... and specify the cutoff frequency,
b = signal.butter(n, Wn, 'high')              # ... build lowpass filter.
spks = signal.filtfilt(b[0], b[1], float_sigs, axis = 0) # Zero-phase filter each trial
```

### Pinch tail

```python
pinch = intanIO(filename = '/Volumes/Seagate Expansion Drive/Intan_Exp/050521_WT_310_Fem_Tx05/WT_310_Fem_Tx05_210505_194540/WT_310_Fem_Tx05_210505_201840.rhd')
```

```python
pinch.parse_header()
```

```python
#could be channel_indexes = [0]
>>> pinch_raw = pinch.get_analogsignal_chunk(block_index=0, seg_index=0, channel_indexes=[9], i_start = 0, i_stop = 250000)
>>> pinch_sigs = pinch.rescale_signal_raw_to_float(pinch_raw, dtype='float64', channel_indexes= [9])
>>> pinch_rate = pinch.get_signal_sampling_rate(channel_indexes= [9])
>>> t_start = pinch.get_signal_t_start(block_index=0, seg_index=0, channel_indexes= [9])
>>> units =pinch.header['signal_channels'][0]['units']
```

```python
fNQ = sampling_rate / 2               # Determine the Nyquist frequency.
n = 10                     # Define the filter order
Wn = 250 / fNQ                  # ... and specify the cutoff frequency,
b = signal.butter(n, Wn, 'low')              # ... build lowpass filter.
pinch_lfp = signal.filtfilt(b[0], b[1], pinch_sigs, axis = 0) # Zero-phase filter each trial
```

```python
plt.figure(figsize = (12, 6))
ax1 = plt.subplot(211)
ax1.plot(pinch_sigs, linewidth = 0.5)
ax1.set_title('Raw Signal')

ax2 = plt.subplot(212)
ax2.plot(pinch_lfp, linewidth = 1)
ax2.set_title('LFP')

plt.tight_layout()
```

### Basal vs Pinch

```python
plt.figure(figsize = (12, 6))
ax1 = plt.subplot(211)
ax1.plot(lfp, linewidth = 1)
ax1.set_title('Control')

ax2 = plt.subplot(212)
ax2.plot(pinch_lfp, linewidth = 1)
ax2.set_title('Pinch')

plt.tight_layout()
```

```python
sf = 25000
win = 4 * sf
freqs, psd = signal.welch(lfp, sf, nperseg=win, axis=0)
p_freqs, p_psd = signal.welch(pinch_lfp, sf, nperseg=win, axis= 0)
```

```python
# Plot the power spectrum
plt.figure(figsize=(10, 4))
plt.plot(freqs, psd, lw=2, label = 'Ctl')
plt.plot(p_freqs, p_psd, lw=2, label = 'Pinch')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Power spectral density (V^2 / Hz)')
plt.title("Welch's periodogram")
plt.xlim([1, 50])
plt.minorticks_on()
plt.legend()
```
