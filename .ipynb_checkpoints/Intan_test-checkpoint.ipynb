### Filtering
#### Low-pass

```python
fNQ = sampling_rate / 2               # Determine the Nyquist frequency.
n = 10                     # Define the filter order
Wn = 350 / fNQ                  # ... and specify the cutoff frequency,
b = signal.butter(n, Wn, 'low')              # ... build lowpass filter.
lfp = signal.filtfilt(b[0], b[1], float_sigs, axis = 0) # Zero-phase filter each trial
```

#### High-pass

```python
fNQ = sampling_rate / 2               # Determine the Nyquist frequency.
n = 10                     # Define the filter order
Wn = 350 / fNQ                  # ... and specify the cutoff frequency,
b = signal.butter(n, Wn, 'high')              # ... build lowpass filter.
spks = signal.filtfilt(b[0], b[1], float_sigs, axis = 0) # Zero-phase filter each trial
```

```python
signal_prep = plt.figure(figsize = (10, 5))
ax1 = plt.subplot(311)
ax1.plot(float_sigs, linewidth = 0.5)
ax1.set_title('Raw Signal')

ax2 = plt.subplot(312)
ax2.plot(lfp, linewidth = 1)
ax2.set_title('LFP')

ax3 = plt.subplot(313)
ax3.plot(spks, linewidth = .5)
ax3.set_title('Spikes')

plt.tight_layout()
```

```python
signal_prep.savefig('/Users/labc02/Documents/PDCB_data/MK-project/Figures/LFP_figures/WT_Fem_Tx05__210625_193654_prep.svg')
```

```python
Fs = sampling_rate               # Define the sampling frequency,
interval = int(Fs)        # ... the interval size,
overlap = int(Fs * 0.95)  # ... and the overlap intervals

                          # Compute the spectrogram
f, t, Sxx = signal.spectrogram(
    lfp.flatten(),                  # Provide the signal,
    fs=Fs,                # ... the sampling frequency,
    nperseg=interval,     # ... the length of a segment,
    noverlap=overlap)     # ... the number of samples to overlap,
```

```python
plt.figure(figsize = (10, 4))
plt.pcolormesh(t, f[0:200], 10 * np.log10(Sxx[0:200,:]),
               cmap='jet')# Plot the result
plt.colorbar()                # ... with a color bar,
plt.ylim(0, 50)              # ... set the frequency range,
plt.minorticks_on()
#plt.xlim(0.5, 20)
plt.xlabel('Time [s]')        # ... and label the axes
plt.ylabel('Frequency [Hz]')
```

```python
sf = 25000
win = 4 * sf
ctl_freqs, ctl_psd = signal.welch(lfp, sf, window = 'hamming', nperseg=win, noverlap = win/2, axis=0)
plt.plot(ctl_freqs, ctl_psd)
plt.xlim(0, 50)
plt.minorticks_on()
```

```python
f_res = cl_fqs[1] - cl_fqs[0]
```

```python
# Defining bands

bands_dic = {'delta': (0.5, 4), 'theta': (4, 12), 'beta': (
    15, 30), 'gamma_slow': (30, 50), 'gamma_fast': (50, 100)}

spectral_dic = {'Date': [], 'Genotype': [], 'Sex': [], 'Tx': [], 'Condition': [], 'total_pw': [], 'delta_abs': [], 'theta_abs': [], 'beta_abs': [],
                'gamma_slow_abs': [], 'gamma_fast_abs': [], 'delta_rel': [], 'theta_rel': [], 'beta_rel': [], 'gamma_slow_rel': [], 'gamma_fast_rel': []}

```

```python
spectral_dic['Date'].append(meta_list[0])
spectral_dic['Genotype'].append(meta_list[1])
spectral_dic['Sex'].append(meta_list[2])
spectral_dic['Tx'].append(meta_list[3])
spectral_dic['Condition'].append(meta_list[4])

total_power = np.round(simps(cl_psd.flatten(), dx = f_res), 3)
spectral_dic['total_pw'].append(total_power)

for band_key in bands_dic:
        limits = bands_dic[band_key]
        band = np.logical_and(cl_fqs >= limits[0], cl_fqs <= limits[1])
        band_abs = np.round(simps(cl_psd.flatten()[band], dx=f_res), 3)
        spectral_dic[band_key + '_abs'].append(band_abs)
        band_rel = np.round(band_abs / total_power, 3)
        spectral_dic[band_key + '_rel'].append(band_rel)
        
spectral_dic['Date'].append(p_meta_list[0])
spectral_dic['Genotype'].append(p_meta_list[1])
spectral_dic['Sex'].append(p_meta_list[2])
spectral_dic['Tx'].append(p_meta_list[3])
spectral_dic['Condition'].append(p_meta_list[4])

total_power = np.round(simps(pl_psd.flatten(), dx = f_res), 3)
spectral_dic['total_pw'].append(total_power)

for band_key in bands_dic:
        limits = bands_dic[band_key]
        band = np.logical_and(cl_fqs >= limits[0], cl_fqs <= limits[1])
        band_abs = np.round(simps(pl_psd.flatten()[band], dx=f_res), 3)
        spectral_dic[band_key + '_abs'].append(band_abs)
        band_rel = np.round(band_abs / total_power, 3)
        spectral_dic[band_key + '_rel'].append(band_rel)
```

```python
spectral_df = pd.DataFrame(spectral_dic)
```
