### Analysis of Auditory Evoked Potentials and Sensory Gating Function


```python
from matplotlib import pyplot as plt
from neo.rawio import IntanRawIO as intanIO
import scipy.signal as sig
import pandas as pd
import numpy as np
from scipy.integrate import simps
import os
```

#### Define probe channels per shank

```python
A32 = {'shank_4' : [5, 11, 12, 10, 4, 14, 3, 6], 'shank_3' : [8, 13, 9, 2, 7, 15, 1, 0],         # Shank numbering is right most to left
           'shank_2' : [29, 24, 18, 19, 31, 23, 16, 28], 'shank_1' :[27, 17, 22, 21, 20, 26, 25, 30]} # That is from medial to lateral
```

#### Select directory of recording and sweeps to analyse (140)
#### Ignore first 5 and get last 145

```python
new_dir = '/Volumes/Seagate Expansion Drive/Intan_Exp/090821_KO_54_6_Fem_Pl/KO_Fem_Pl_Gating__210809_195208'

print(len(os.listdir(new_dir)))

soi = os.listdir(new_dir)[5:145]
```

#### Get sweep signal

```python
sweep = new_dir + soi[0]
reader = intanIO(filename = sweep)
```

```python
# low pass filter
fNQ = 25000 / 2               # Determine the Nyquist frequency.
n = 10                     # Define the filter order
Wn = 200 / fNQ                  # ... and specify the cutoff frequency,
b = sig.butter(n, Wn, 'low')              # ... build lowpass filter.
```

```python
time_roi = np.arange(-0.1, (27500/25000)-0.1, 1/25000)
```

```python
shank_mat = np.zeros((8, 150, 27500))

for file_idx, file in enumerate(soi):
    sweep = new_dir +'/' + file
    reader = intanIO(filename = sweep)
    reader.parse_header()
    size = reader.get_signal_size(block_index=0, seg_index=0,channel_indexes = [0]) # get the size of recording
    analog_1 = reader.get_analogsignal_chunk(block_index=0, seg_index=0,channel_indexes=[35], i_start = 0, i_stop = size) # read analog_in
    analog_scaled = reader.rescale_signal_raw_to_float(analog_1, dtype='float64', channel_indexes= [35])
    trigger_id, _ = sig.find_peaks(analog_scaled.flatten(), height = 1.5) # find trigger signal
    trigger_id += 625
    roi = (trigger_id[0]-2500, trigger_id[0]+25000) # region of interest within recording (100 ms before trigger to 1 sec after)

    for ch_idx, channel in enumerate(A32['shank_2']):
        raw_sig = reader.get_analogsignal_chunk(block_index=0, seg_index=0,channel_indexes=[channel], i_start = roi[0], i_stop = roi[1]) # read analog_in
        sig_scaled = reader.rescale_signal_raw_to_float(raw_sig, dtype='float64', channel_indexes= [channel])
        lfp = sig.filtfilt(b[0], b[1], sig_scaled, axis = 0) # Zero-phase filter each trial
        
        shank_mat[ch_idx, file_idx, :] = lfp.flatten()-lfp.flatten().mean()

shank_mean = shank_mat.mean(1)
```

```python
fig, axs = plt.subplots(nrows = 8, ncols = 1, sharex = True, figsize = (10, 8))
for ax_id, ax in enumerate(axs):
    ax.plot(time_roi, shank_mean[ax_id, :], linewidth = 1)
    ax.vlines(x = 0.0, ymin = -25, ymax = 25, colors='r', linestyles='dotted')
    ax.vlines(x = 0.5, ymin = -25, ymax = 25, colors='r', linestyles='dotted')
    ax.spines['right'].set_visible(False)
    ax.spines['top'].set_visible(False)
    ax.set_xticks(ticks = np.arange(0.0,1.1,0.05), minor = True)
```

```python
CS_mat = np.zeros((4, 8, 150, 5000))

for file_idx, file in enumerate(soi):
    #print(f'File: {file_idx}')
    sweep = new_dir +'/'+ file
    reader = intanIO(filename = sweep)
    reader.parse_header()
    size = reader.get_signal_size(block_index=0, seg_index=0,channel_indexes = [0]) # get the size of recording
    analog_1 = reader.get_analogsignal_chunk(block_index=0, seg_index=0,channel_indexes=[35], i_start = 0, i_stop = size) # read analog_in
    analog_scaled = reader.rescale_signal_raw_to_float(analog_1, dtype='float64', channel_indexes= [35])
    trigger_id, _ = sig.find_peaks(analog_scaled.flatten(), height = 1.5) # find trigger signal
    trigger_id += 625
    roi = (trigger_id[0], trigger_id[0]+5000) # region of interest within recording (100 ms before trigger to 0.2 sec after)

    for shank in A32:
        sh_idx = int(shank[-1])-1
        for ch_idx, channel in enumerate(A32[shank]):
            raw_sig = reader.get_analogsignal_chunk(block_index=0, seg_index=0,channel_indexes=[channel], i_start = roi[0], i_stop = roi[1]) # read analog_in
            sig_scaled = reader.rescale_signal_raw_to_float(raw_sig, dtype='float64', channel_indexes= [channel])
            lfp = sig.filtfilt(b[0], b[1], sig_scaled, axis = 0) # Zero-phase filter each trial

            CS_mat[sh_idx, ch_idx, file_idx, :] = lfp.flatten()-lfp.flatten().mean()

CS_mean = CS_mat.mean(2)
```

```python
CS_fig, axs = plt.subplots(nrows = 8, ncols = 4, figsize = (8, 5))
for column in range(4):
    for row in range(8):
        axs[row, column].plot(CS_mean[column, row])
        axs[row, column].set_axis_off()
plt.annotate('CA1', (0.92,0.87), xycoords = 'figure fraction', fontsize = 12)
plt.annotate('CA3', (0.92,0.23), xycoords = 'figure fraction', fontsize = 12)
plt.annotate('Medial', (0.79,0.05), xycoords = 'figure fraction', fontsize = 12)
plt.annotate('Lateral', (0.1,0.05), xycoords = 'figure fraction', fontsize = 12)
```

```python
TS_mat = np.zeros((4, 8, 150, 5000))

for file_idx, file in enumerate(soi):
    #print(f'File: {file_idx}')
    sweep = new_dir +'/'+ file
    reader = intanIO(filename = sweep)
    reader.parse_header()
    size = reader.get_signal_size(block_index=0, seg_index=0,channel_indexes = [0]) # get the size of recording
    analog_1 = reader.get_analogsignal_chunk(block_index=0, seg_index=0,channel_indexes=[35], i_start = 0, i_stop = size) # read analog_in
    analog_scaled = reader.rescale_signal_raw_to_float(analog_1, dtype='float64', channel_indexes= [35])
    trigger_id, _ = sig.find_peaks(analog_scaled.flatten(), height = 1.5) # find trigger signal
    trigger_id += 13125
    roi = (trigger_id[0], trigger_id[0]+5000) # region of interest within recording (100 ms before trigger to 0.2 sec after)

    for shank in A32:
        sh_idx = int(shank[-1])-1
        for ch_idx, channel in enumerate(A32[shank]):
            raw_sig = reader.get_analogsignal_chunk(block_index=0, seg_index=0,channel_indexes=[channel], i_start = roi[0], i_stop = roi[1]) # read analog_in
            sig_scaled = reader.rescale_signal_raw_to_float(raw_sig, dtype='float64', channel_indexes= [channel])
            lfp = sig.filtfilt(b[0], b[1], sig_scaled, axis = 0) # Zero-phase filter each trial

            TS_mat[sh_idx, ch_idx, file_idx, :] = lfp.flatten()-lfp.flatten().mean()

TS_mean = TS_mat.mean(2)
```

```python
TS_fig, axs = plt.subplots(nrows = 8, ncols = 4, figsize = (8, 5))
for column in range(4):
    for row in range(8):
        axs[row, column].plot(TS_mean[column, row])
        axs[row, column].set_axis_off()
plt.annotate('CA1', (0.92,0.87), xycoords = 'figure fraction', fontsize = 12)
plt.annotate('CA3', (0.92,0.23), xycoords = 'figure fraction', fontsize = 12)
plt.annotate('Medial', (0.79,0.05), xycoords = 'figure fraction', fontsize = 12)
plt.annotate('Lateral', (0.1,0.05), xycoords = 'figure fraction', fontsize = 12)
```
